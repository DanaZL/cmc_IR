```python
with open("./lordring.txt") as f:
    text = [] #a list of words
    for line in f:
        tokens = re.findall('[\w]+',line.decode("utf-8").lower(), re.U)        
        text += tokens
#     for idx, token in tqdm(enumerate(text)):
#         text[idx] = RusLem.parse(token)[0].normal_form
    for i, token in tqdm(enumerate(text)):
        text[i] = RusLem.parse(token)[0].normal_form
```

```python
bprint(text[:100], sep=" ")
```

```python
token_freqs = {}
for token in tqdm(text):
    if token not in token_freqs:
        token_freqs[token] = 0
    token_freqs[token] += 1
```

```python
sorted_by_freq = sorted(token_freqs.items(), key=operator.itemgetter(1), reverse=True)
for token in sorted_by_freq:
    print "%s\t%s" % (token[0], token[1])
```
